#  PART 1: AI Concepts + Dataset Creation + Data Understanding

```python
# ============================================================
# PROJECT : Stock Price Predictor
# PART 1 : AI Concepts + Dataset Creation + Data Understanding
# ============================================================

# ============================================================
# Import Required Libraries
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Machine Learning Libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

print("=" * 65)
print("           STOCK PRICE PREDICTOR PROJECT")
print("=" * 65)

print("\nProject Started Successfully!\n")

# ============================================================
# AI Concepts
# ============================================================

print("What is Artificial Intelligence (AI)?")
print("Artificial Intelligence enables machines to perform tasks")
print("that normally require human intelligence.\n")

print("What is Machine Learning (ML)?")
print("Machine Learning is a branch of AI where computers")
print("learn patterns from historical data.\n")

print("About This Project")
print("This project predicts the Closing Stock Price")
print("using historical stock market data.\n")

# ============================================================
# Creating Custom Dataset
# (No CSV File Required)
# ============================================================

stock_data = {

    "Day": list(range(1,31)),

    "Open":[
        100,102,104,103,105,106,108,110,109,111,
        113,114,116,117,119,118,120,122,121,123,
        125,126,128,129,130,132,133,135,136,138
    ],

    "High":[
        104,105,107,106,108,109,111,113,112,114,
        116,118,119,120,122,121,123,125,124,126,
        128,129,131,132,133,135,136,138,139,141
    ],

    "Low":[
        98,100,102,101,103,104,106,108,107,109,
        111,112,114,115,117,116,118,120,119,121,
        123,124,126,127,128,130,131,133,134,136
    ],

    "Volume":[
        5000,5200,5100,5300,5400,
        5500,5600,5700,5800,5900,
        6000,6100,6200,6300,6400,
        6500,6600,6700,6800,6900,
        7000,7100,7200,7300,7400,
        7500,7600,7700,7800,7900
    ],

    "Close":[
        102,104,105,104,107,
        108,110,111,110,113,
        115,116,118,119,120,
        119,122,123,122,124,
        126,127,129,130,131,
        133,134,136,137,140
    ]

}

# Convert Dictionary into DataFrame

stocks = pd.DataFrame(stock_data)

# ============================================================
# Display Dataset
# ============================================================

print("="*65)
print("FIRST FIVE RECORDS")
print("="*65)

print(stocks.head())

# ============================================================
# Dataset Shape
# ============================================================

print("\nDataset Shape")

print(stocks.shape)

# ============================================================
# Column Names
# ============================================================

print("\nColumn Names")

print(stocks.columns)

# ============================================================
# Dataset Information
# ============================================================

print("\nDataset Information")

print(stocks.info())

# ============================================================
# Statistical Summary
# ============================================================

print("\nStatistical Summary")

print(stocks.describe())

# ============================================================
# Missing Values
# ============================================================

print("\nMissing Values")

print(stocks.isnull().sum())

# ============================================================
# Duplicate Records
# ============================================================

print("\nDuplicate Records")

print(stocks.duplicated().sum())

# ============================================================
# Dataset Summary
# ============================================================

print("\n" + "="*65)
print("DATASET SUMMARY")
print("="*65)

print("Total Records :", len(stocks))
print("Total Columns :", len(stocks.columns))

print("\nFeatures Used")

for column in stocks.columns:
    print("-", column)

# ============================================================
# Feature Engineering
# ============================================================

# Daily Price Change

stocks["Price_Change"] = stocks["Close"] - stocks["Open"]

# High-Low Difference

stocks["High_Low_Diff"] = stocks["High"] - stocks["Low"]

print("\nDataset After Feature Engineering")

print(stocks.head())

# ============================================================
# Average Values
# ============================================================

print("\nAverage Opening Price")

print(stocks["Open"].mean())

print("\nAverage Closing Price")

print(stocks["Close"].mean())

print("\nAverage Trading Volume")

print(stocks["Volume"].mean())

# ============================================================
# Highest and Lowest Closing Price
# ============================================================

print("\nHighest Closing Price")

print(stocks["Close"].max())

print("\nLowest Closing Price")

print(stocks["Close"].min())

print("\nPart 1 Completed Successfully!")
```

#  PART 2: Data Preprocessing and Data Visualization

```python
# ============================================================
# PART 2 : DATA PREPROCESSING AND VISUALIZATION
# ============================================================

print("=" * 65)
print("PART 2 : DATA PREPROCESSING AND VISUALIZATION")
print("=" * 65)

# ============================================================
# Display Updated Dataset
# ============================================================

print("\nFirst Five Records")
print(stocks.head())

# ============================================================
# Check Data Types
# ============================================================

print("\nData Types")
print(stocks.dtypes)

# ============================================================
# Check Missing Values
# ============================================================

print("\nMissing Values")
print(stocks.isnull().sum())

# ============================================================
# Check Duplicate Records
# ============================================================

print("\nDuplicate Records")
print(stocks.duplicated().sum())

# ============================================================
# Statistical Summary
# ============================================================

print("\nStatistical Summary")
print(stocks.describe())

# ============================================================
# Line Chart - Closing Price
# ============================================================

plt.figure(figsize=(10,5))

plt.plot(
    stocks["Day"],
    stocks["Close"],
    marker="o"
)

plt.title("Closing Stock Price")
plt.xlabel("Day")
plt.ylabel("Closing Price")
plt.grid(True)

plt.show()

# ============================================================
# Open Price vs Close Price
# ============================================================

plt.figure(figsize=(10,5))

plt.plot(
    stocks["Day"],
    stocks["Open"],
    marker="o",
    label="Open Price"
)

plt.plot(
    stocks["Day"],
    stocks["Close"],
    marker="s",
    label="Close Price"
)

plt.title("Open Price vs Close Price")
plt.xlabel("Day")
plt.ylabel("Price")

plt.legend()

plt.grid(True)

plt.show()

# ============================================================
# Volume Bar Chart
# ============================================================

plt.figure(figsize=(10,5))

plt.bar(
    stocks["Day"],
    stocks["Volume"]
)

plt.title("Daily Trading Volume")
plt.xlabel("Day")
plt.ylabel("Volume")

plt.show()

# ============================================================
# Histogram
# ============================================================

plt.figure(figsize=(8,5))

plt.hist(
    stocks["Close"],
    bins=8,
    edgecolor="black"
)

plt.title("Distribution of Closing Price")
plt.xlabel("Closing Price")
plt.ylabel("Frequency")

plt.show()

# ============================================================
# Box Plot
# ============================================================

plt.figure(figsize=(6,5))

plt.boxplot(stocks["Close"])

plt.title("Box Plot of Closing Price")
plt.ylabel("Closing Price")

plt.show()

# ============================================================
# Scatter Plot
# ============================================================

plt.figure(figsize=(8,5))

plt.scatter(
    stocks["Open"],
    stocks["Close"]
)

plt.title("Open Price vs Close Price")
plt.xlabel("Open Price")
plt.ylabel("Close Price")

plt.grid(True)

plt.show()

# ============================================================
# Correlation Matrix
# ============================================================

print("\nCorrelation Matrix")

correlation = stocks.corr(numeric_only=True)

print(correlation)

# ============================================================
# Feature Selection
# ============================================================

print("\nPreparing Features and Target Variable...")

X = stocks[[
    "Open",
    "High",
    "Low",
    "Volume"
]]

y = stocks["Close"]

print("\nFeatures (X)")
print(X.head())

print("\nTarget (y)")
print(y.head())

# ============================================================
# Feature Information
# ============================================================

print("\nNumber of Features :", X.shape[1])

print("Number of Records :", X.shape[0])

print("\nFeature Names")

for column in X.columns:
    print("-", column)

print("\nTarget Column")
print("Close")

# ============================================================
# Average Values
# ============================================================

print("\nAverage Open Price :", stocks["Open"].mean())
print("Average Close Price :", stocks["Close"].mean())
print("Average Volume :", stocks["Volume"].mean())

print("\nPart 2 Completed Successfully!")
```

# PART 3: Linear Regression Model and Evaluation

```python
# ============================================================
# PART 3 : MACHINE LEARNING MODEL
# ============================================================

print("=" * 65)
print("PART 3 : MACHINE LEARNING MODEL")
print("=" * 65)

# ============================================================
# Split Dataset
# ============================================================

print("\nSplitting Dataset into Training and Testing Data...")

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("\nTraining Data Shape :", X_train.shape)
print("Testing Data Shape  :", X_test.shape)

# ============================================================
# Create Linear Regression Model
# ============================================================

print("\nCreating Linear Regression Model...")

model = LinearRegression()

# ============================================================
# Train Model
# ============================================================

print("Training Model...")

model.fit(X_train, y_train)

print("Model Trained Successfully!")

# ============================================================
# Model Coefficients
# ============================================================

print("\nModel Coefficients")

for feature, coefficient in zip(X.columns, model.coef_):
    print(f"{feature:10} : {coefficient:.4f}")

print(f"\nIntercept : {model.intercept_:.4f}")

# ============================================================
# Predict Stock Prices
# ============================================================

print("\nPredicting Stock Prices...")

y_pred = model.predict(X_test)

print("Prediction Completed!")

# ============================================================
# Actual vs Predicted Values
# ============================================================

results = pd.DataFrame({
    "Actual Price": y_test.values,
    "Predicted Price": np.round(y_pred, 2)
})

print("\nActual vs Predicted Stock Prices")

print(results)

# ============================================================
# Model Evaluation
# ============================================================

print("\n" + "="*50)
print("MODEL EVALUATION")
print("="*50)

r2 = r2_score(y_test, y_pred)

mae = mean_absolute_error(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

print(f"R² Score               : {r2:.4f}")
print(f"Mean Absolute Error    : {mae:.4f}")
print(f"Mean Squared Error     : {mse:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")

# ============================================================
# Training and Testing Scores
# ============================================================

train_score = model.score(X_train, y_train)

test_score = model.score(X_test, y_test)

print("\nTraining Score :", round(train_score*100,2), "%")

print("Testing Score  :", round(test_score*100,2), "%")

# ============================================================
# Actual vs Predicted Graph
# ============================================================

plt.figure(figsize=(8,5))

plt.plot(
    y_test.values,
    marker="o",
    label="Actual Price"
)

plt.plot(
    y_pred,
    marker="s",
    label="Predicted Price"
)

plt.title("Actual Price vs Predicted Price")
plt.xlabel("Test Data")
plt.ylabel("Closing Price")

plt.legend()

plt.grid(True)

plt.show()

# ============================================================
# Scatter Plot
# ============================================================

plt.figure(figsize=(6,6))

plt.scatter(
    y_test,
    y_pred
)

plt.plot(
    [y.min(), y.max()],
    [y.min(), y.max()]
)

plt.title("Actual vs Predicted Price")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")

plt.grid(True)

plt.show()

# ============================================================
# Residual Errors
# ============================================================

residual = y_test.values - y_pred

plt.figure(figsize=(8,5))

plt.bar(
    range(len(residual)),
    residual
)

plt.title("Prediction Error")
plt.xlabel("Test Samples")
plt.ylabel("Error")

plt.show()

# ============================================================
# Model Summary
# ============================================================

print("\n" + "="*65)
print("MODEL SUMMARY")
print("="*65)

print("Algorithm Used      : Linear Regression")
print("Learning Type       : Supervised Learning")
print("Prediction Target   : Closing Stock Price")
print("Training Samples    :", X_train.shape[0])
print("Testing Samples     :", X_test.shape[0])
print(f"R² Score            : {r2:.4f}")

print("\nMachine Learning Model Completed Successfully!")
```
